```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"D:\mydata\Elevate Labs\Titanic-Dataset.csv")
```

```python
# ------------------------------
# 1️⃣ Summary Statistics
# ------------------------------
print("=== Dataset Information ===")
print(df.info())

print("\n=== Summary Statistics ===")
print(df.describe(include='all'))

# Check missing values
print(df.isnull().sum())
```

```python
# ------------------------------
# 2️⃣ Histograms & Boxplots
# ------------------------------

# Select only numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Histograms for numeric features
df[numeric_cols].hist(figsize=(12, 10), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numeric Features", fontsize=16)
plt.tight_layout()
plt.show()

# Boxplots for numeric features
plt.figure(figsize=(12, 8))
df[numeric_cols].boxplot()
plt.title("Boxplots of Numeric Features")
plt.xticks(rotation=45)
plt.show()
```

```python
# ------------------------------
# 3️⃣ Pairplot & Correlation Matrix
# ------------------------------

# Correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# Pairplot (use a sample if dataset is large)
sns.pairplot(df[numeric_cols])
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()
```

```python
# ------------------------------
# 4️⃣ Identify Patterns / Trends
# ------------------------------

# Example: Survival rate by gender (if 'Survived' column exists)
if 'Survived' in df.columns and 'Sex' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.barplot(x='Sex', y='Survived', data=df, hue='Sex', palette='Set2', legend=False)

    plt.title("Survival Rate by Gender")
    plt.show()

# Example: Survival rate by passenger class
if 'Survived' in df.columns and 'Pclass' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.barplot(x='Sex', y='Survived', data=df, hue='Sex', palette='Set2', legend=False)

    plt.title("Survival Rate by Passenger Class")
    plt.show()
```

```python
# ------------------------------
# 5️⃣ Feature-level Inferences
# ------------------------------

print("\n=== Basic Insights ===")
if 'Age' in df.columns:
    print(f"Average passenger age: {df['Age'].mean():.2f}")
    print(f"Oldest passenger age: {df['Age'].max()}")
    print(f"Youngest passenger age: {df['Age'].min()}")

if 'Fare' in df.columns:
    print(f"Average fare paid: {df['Fare'].mean():.2f}")
    print(f"Highest fare: {df['Fare'].max():.2f}")
    print(f"Lowest fare: {df['Fare'].min():.2f}")

if 'Survived' in df.columns:
    survival_rate = df['Survived'].mean() * 100
    print(f"Overall survival rate: {survival_rate:.2f}%")


```
